# Advanced Sales Data Analysis

A full pipeline: data cleaning, EDA, KPI calculations, and insights.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

# Load raw data
df = pd.read_csv('../data/raw_sales_data.csv')
df.head()
```

## Step 1: Clean the data

```python
# Drop rows with missing date
df = df.dropna(subset=['Date'])

# Clean column names
df.columns = df.columns.str.strip()

# Remove duplicates
df = df.drop_duplicates()

# Replace empty or NaN regions/products
df['Region'] = df['Region'].fillna('Unknown')
df['Product'] = df['Product'].replace('', 'Unknown').fillna('Unknown')

# Drop rows with missing numerical values
df = df.dropna(subset=['Units Sold', 'Unit Price'])

# Convert data types
df['Date'] = pd.to_datetime(df['Date'])
df['Units Sold'] = df['Units Sold'].astype(int)
df['Unit Price'] = df['Unit Price'].astype(float)

# Remove obvious outliers
df = df[df['Units Sold'] < 5000]

# Add calculated revenue
df['Revenue'] = df['Units Sold'] * df['Unit Price']
df.head()
```

## Step 2: Product categorization

```python
# Create a product category mapping
category_map = {
    'Laptop': 'Tech',
    'Tablet': 'Tech',
    'Phone': 'Mobile',
    'Monitor': 'Accessories',
    'Unknown': 'Other'
}
df['Category'] = df['Product'].map(category_map).fillna('Other')
df['Month'] = df['Date'].dt.to_period('M')
df.head()
```

## Step 3: EDA â€“ Trends and KPIs

```python
# Revenue by Region
sns.barplot(data=df, x='Region', y='Revenue', estimator=sum)
plt.title('Total Revenue by Region')
plt.show()

# Revenue by Month
monthly_revenue = df.groupby('Month')['Revenue'].sum()
monthly_revenue.plot(marker='o', figsize=(8,4))
plt.title('Monthly Revenue')
plt.ylabel('Revenue')
plt.grid(True)
plt.show()

# Revenue by Category
sns.barplot(data=df, x='Category', y='Revenue', estimator=sum)
plt.title('Revenue by Product Category')
plt.show()
```

## Step 4: Business KPIs

```python
kpi = {
    'Total Revenue': df['Revenue'].sum(),
    'Total Units Sold': df['Units Sold'].sum(),
    'Average Order Value': df['Revenue'].sum() / len(df),
    'Revenue per Unit': df['Revenue'].sum() / df['Units Sold'].sum(),
    'Unique Products': df['Product'].nunique(),
    'Unique Regions': df['Region'].nunique()
}
for key, value in kpi.items():
    print(f"{key}: {value:,.2f}")
```

## Step 5: Forecasting Monthly Revenue (Linear Regression)

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import matplotlib.dates as mdates

# Ð“Ñ€ÑƒÐ¿ÑƒÐ²Ð°Ð½Ð½Ñ Ð¿Ð¾ Ð¼Ñ–ÑÑÑ†ÑÑ…
monthly_rev = df.groupby(df['Date'].dt.to_period('M'))['Revenue'].sum().reset_index()
monthly_rev['Month'] = monthly_rev['Date'].dt.to_timestamp()
monthly_rev['MonthIndex'] = np.arange(len(monthly_rev))

# ÐœÐ¾Ð´ÐµÐ»ÑŒ
X = monthly_rev[['MonthIndex']]
y = monthly_rev['Revenue']
model = LinearRegression()
model.fit(X, y)

# ÐŸÑ€Ð¾Ð³Ð½Ð¾Ð· Ð½Ð° 3 Ð¼Ð°Ð¹Ð±ÑƒÑ‚Ð½Ñ– Ð¼Ñ–ÑÑÑ†Ñ–
future_index = np.arange(len(monthly_rev), len(monthly_rev) + 3)
future_pred = model.predict(future_index.reshape(-1, 1))
future_months = pd.date_range(start=monthly_rev['Month'].iloc[-1] + pd.offsets.MonthBegin(), periods=3, freq='MS')

# Ð’Ñ–Ð·ÑƒÐ°Ð»Ñ–Ð·Ð°Ñ†Ñ–Ñ
plt.figure(figsize=(10,5))
plt.plot(monthly_rev['Month'], y, marker='o', label='Actual')
plt.plot(future_months, future_pred, marker='x', linestyle='--', label='Forecast')
plt.title('Monthly Revenue Forecast')
plt.ylabel('Revenue')
plt.xlabel('Month')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
